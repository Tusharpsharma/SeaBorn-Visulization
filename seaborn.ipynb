line plot graph

A line graph (or line plot) is a type of data visualization that displays information as a series of data points called 'markers' connected by straight line segments. Line graphs are commonly used to visualize trends over time or to display data points in relation to one another.

```python
# print(tips)  #hue like by male or femaale lines with diff colour
# style = diff style lines
#palette - colour style - use any to find diff in error then choose one
sns.lineplot(data =tips, x= 'total_bill', y ='tip',hue='sex',style= 'day', palette='RdBu')
plt.show()
```

Barplot Graph

A bar plot is a common data visualization technique used to compare different categories of data. Each category is represented by a rectangular bar, where the length of the bar is proportional to the value it represents. Bar plots can be vertical or horizontal, and they can display counts, averages, or other summary statistics.

```python
sns.barplot(data =tips, x= 'sex', y ='tip',hue='sex')
plt.show()
```

```python
#color = 'blue also we can choose
#x or y we can put 'who' will give diff options
sns.barplot(data =tips, x= 'sex', y ='tip',hue='sex',palette='dark:blue',saturation=0.2)
plt.show()
```

```python
from numpy import mean ##or  median
data = sns.load_dataset("titanic")
# ci = False will remove median line
#facecolor [1,1,1,0],edgecolor =0.2
#order will give order of bars
sns.barplot(x="who", y="survived", data=data,estimator=mean, order=['child','woman','man'],color='red',edgecolor= 'black')

plt.show()
```

Scatter Plot


A scatter plot is a type of data visualization that uses dots to represent the values obtained for two different variables—one plotted along the x-axis and the other plotted along the y-axis. Scatter plots are useful for showing relationships, correlations, and distributions between two numerical variables.

```python
# Marker will give different pointer like star,
#style will five different styles 
#size (20.200), alpha 0 to 1 for transpancy
sns.scatterplot(data =tips, x= 'total_bill', y ='tip',hue='sex',marker ='*',alpha =0.8)
plt.show()
# print(tips)
```

Histogram plot

A histogram is a graphical representation that organizes a group of data points into specified ranges (bins). It displays the frequency of data points that fall within each bin, allowing you to see the distribution of the data, including its central tendency, variability, and skewness.

```python
#discrete will remove where data is not there
#kde = true will give trend line
#linestyle will give --- dotted lines
#bins=20 will divide data into 20 lines
sns.histplot(data = tips, x= 'total_bill', discrete= 'False', kde=True, hue='sex', edgecolor='red', linestyle='--', bins=20)
plt.show()
```

violen plot

A violin plot combines aspects of a box plot and a kernel density plot. It shows the distribution of the data across different categories while also indicating the probability density of the data at different values. This makes it easier to understand the distribution of the data along with its central tendency.

```python
# split = 'True will split both'
sns.violinplot(data = tips, x= 'sex', y= 'total_bill', hue='sex', order=['Male', 'Female'])
plt.show()
```

Joint Plot

A joint plot is a data visualization technique that combines scatter plots and histograms to show the relationship between two variables along with the distribution of each variable. It is particularly useful for examining the correlation between two continuous variables and understanding their distributions simultaneously.

```python
#kind = 'reg' will give regression line
sns.jointplot(data = tips, x= 'tip', y= 'total_bill', kind='reg')
plt.show()
```

Pair Plot


A pair plot is a powerful visualization tool that allows you to see the relationships between multiple variables in a dataset. It creates a grid of scatter plots for each pair of numerical variables in the dataset, along with histograms or density plots on the diagonal for each variable. This helps in understanding the pairwise relationships and distributions in multivariate data.

```python
#will show possible graphs
sns.pairplot(data=tips, hue='sex')
plt.show()
```

Swarm Plot

A swarm plot is a type of categorical scatter plot that shows the distribution of data points for one or more categories. Unlike a traditional scatter plot, where points can overlap, swarm plots adjust the positions of points along the categorical axis to avoid overlap. This allows you to see the density and distribution of points in each category clearly.

```python
sns.swarmplot(data = tips, x= 'day', y= 'total_bill',hue = 'sex', palette='husl')
plt.show()
```

Count Plot

A count plot is a type of bar plot that shows the number of occurrences (counts) of each category in a categorical variable. It’s useful for visualizing the frequency distribution of categories in a dataset, allowing you to quickly see which categories are most or least common.

```python
#count plot in Seaborn is used to display the count of observations in each category.
tips = sns.load_dataset("tips")

sns.countplot(data=tips, x='day', hue='sex')
```

Strip Plot

A strip plot is a type of scatter plot that displays the values of a quantitative variable for different categories. Unlike a traditional scatter plot, strip plots are particularly useful for visualizing the distribution of data points in categorical variables, as they spread the points along the categorical axis to avoid overlap. This helps to reveal the underlying distribution of the data within each category.

```python

# A strip plot in Seaborn is used to display the distribution of data points along a categorical axis. 
# It is similar to a swarm plot, but the points in a strip plot can overlap
 
# Jitter = true when same data points are there to spread them
#dodge = true parameter separates the data points for different subgroups along the categorical axis, when using the
#can use marker(*)
#sixe = 20

sns.stripplot(data = tips, x= 'day', y= 'total_bill',hue = 'sex', palette='husl', dodge='True', jitter=True)
plt.show()
```

BOX Plot

A box plot (or whisker plot) is a standardized way of displaying the distribution of a dataset based on a five-number summary: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It visually represents the central tendency, variability, and potential outliers in the data. Box plots are particularly useful for comparing distributions between multiple groups or categories

```python
#A box plot (or box-and-whisker plot) is used to display the distribution of a dataset and provide 
# summary statistics such as the median, quartiles, and outliers.

#orient = y or h we can use for axes orientation

sns. boxplot(data = tips, x= 'day', y= 'total_bill',hue='sex',palette='husl')
plt.show()
```

Heat Map


A heat map is a data visualization technique that shows the magnitude of a phenomenon as color in two dimensions. The values are represented in a matrix format, where the color intensity reflects the value of the data. Heat maps are particularly useful for displaying complex data sets and for identifying patterns, correlations, and areas of concentration.

```python

tips_df = tips[['total_bill', 'tip', 'size']]
tips_df.head()


```

```python

# annot=True: Displays the correlation coefficients on the heat map.
# cmap='coolwarm': Specifies the color palette (cool to warm).
# fmt=".2f": Formats the annotations to two decimal places.


sns.heatmap(data=tips_df.corr(), annot=True)
```

Cat Plot

A cat plot in Seaborn is a versatile function that combines several categorical plots into one. It's a convenient way to visualize relationships between categorical and numerical variables while allowing for various types of visualizations, such as box plots, violin plots, or bar plots, based on the input parameters. The catplot() function is especially useful for exploring the distribution of a numerical variable across different categories.

```python
#A cat plot in Seaborn is a versatile function that combines several categorical plots into one interface. 
# It allows you to visualize categorical data using different types of plots, 
# such as box plots, strip plots, swarm plots, and point plots

#Inside Kind we can put box,swamp,strip, poin plot it will create that king of graph#

tips = sns.load_dataset("tips")

# Create a cat plot using a box plot
sns.catplot(data=tips, x='day', y='total_bill', kind='box')

plt.title('Total Bill by Day')
plt.show()
```
